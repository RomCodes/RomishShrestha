Rolling avg:

means you take an average of values for every x amount of time 

```python
#get rolling avg. window param means how many rows to take avg from
df['Close'].plot(figsize = (17,10))
df.rolling(window= 60).mean()['Close'].plot(figsize = (17,10))

#first six values are all nan because there arent 7 values yet to take rolling avg
```

```python
#rolling avg with label
df['Close : 30 day mean'] = df['Close'].rolling(window =30).mean()
```

```python
df['Close'].plot()
ax = df['Close : 30 day mean'].plot(figsize= (17,10))
ax.legend()
```

```python
#avg of everything that came before it rather than rolling avg of a period of time
#expanding is average of everything that came before without a window of avg 

df['Close'].expanding().mean().plot(figsize=(17,10))
```
